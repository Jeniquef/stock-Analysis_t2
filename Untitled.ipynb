```python
!pip install yfinance
import yfinance as yf
import matplotlib.pyplot as plt
import pandas as pd
from scipy import stats
```

```python
# Download data for Apple, Samsung, and the NASDAQ-100 Technology Index
apple = yf.download('AAPL', start='2018-01-01', end='2023-12-31')
samsung = yf.download('005930.KS', start='2018-01-01', end='2023-12-31')
nasdaq_tech = yf.download('^NDXT', start='2018-01-01', end='2023-12-31')
apple.head()
```

```python
# Calculate daily returns
apple['Daily Return'] = apple['Adj Close'].pct_change()
samsung['Daily Return'] = samsung['Adj Close'].pct_change()
nasdaq_tech['Daily Return'] = nasdaq_tech['Adj Close'].pct_change()
```

```python
apple[['Adj Close', 'Daily Return']].head()
samsung[['Adj Close', 'Daily Return']].head()
nasdaq_tech[['Adj Close', 'Daily Return']].head()
```

```python
# Normalize prices for comparison
apple['Normalized Price'] = apple['Adj Close'] / apple['Adj Close'].iloc[0]
samsung['Normalized Price'] = samsung['Adj Close'] / samsung['Adj Close'].iloc[0]
nasdaq_tech['Normalized Price'] = nasdaq_tech['Adj Close'] / nasdaq_tech['Adj Close'].iloc[0]
```

```python
# For Jupyter Notebook
apple[['Adj Close', 'Normalized Price']].head()
samsung[['Adj Close', 'Normalized Price']].head()
nasdaq_tech[['Adj Close', 'Normalized Price']].head()
```

```python

# Plot normalized stock prices for comparison
plt.figure(figsize=(10,5))
plt.plot(apple['Normalized Price'], label='Apple')
plt.plot(samsung['Normalized Price'], label='Samsung')
plt.plot(nasdaq_tech['Normalized Price'], label='NASDAQ Tech Index')
plt.title('Stock Price Comparison: Apple, Samsung, and NASDAQ-100 Tech Index')
plt.legend()
plt.show()
```

```python
apple['Volatility'] = apple['Daily Return'].rolling(window=30).std()
samsung['Volatility'] = samsung['Daily Return'].rolling(window=30).std()
nasdaq_tech['Volatility'] = nasdaq_tech['Daily Return'].rolling(window=30).std()

# Plot volatility
plt.figure(figsize=(10,5))
plt.plot(apple['Volatility'], label='Apple Volatility')
plt.plot(samsung['Volatility'], label='Samsung Volatility')
plt.plot(nasdaq_tech['Volatility'], label='NASDAQ Tech Index Volatility')
plt.title('Rolling Volatility Comparison: Apple, Samsung, and NASDAQ-100 Tech Index')
plt.legend()
plt.savefig('my_plot1.png')
plt.show()
```

```python
# Define the stock tickers
tickers = ['AAPL', '005930.KS', 'QQQ']

# Download adjusted closing prices
data = yf.download(tickers, start="2019-01-01", end="2023-01-01")['Adj Close']

# Calculate daily percentage returns
returns = data.pct_change().dropna()
```

```python
# Split into pre-COVID (before March 1, 2020) and post-COVID (after March 1, 2020)
pre_covid_returns = returns.loc[:'2020-03-01']
post_covid_returns = returns.loc['2020-03-01':]
```

```python
# Split into pre-COVID (before March 1, 2020) and post-COVID (after March 1, 2020)
pre_covid_returns = returns.loc[:'2020-03-01']
post_covid_returns = returns.loc['2020-03-01':]
```

```python
# Pre-COVID t-tests

# Apple vs Tech Sector (QQQ)
t_stat_apple_pre, p_value_apple_pre = stats.ttest_ind(pre_covid_returns['AAPL'], pre_covid_returns['QQQ'])

# Samsung vs Tech Sector (QQQ)
t_stat_samsung_pre, p_value_samsung_pre = stats.ttest_ind(pre_covid_returns['005930.KS'], pre_covid_returns['QQQ'])

print(f"Pre-COVID Apple vs Tech Sector: T-statistic = {t_stat_apple_pre}, P-value = {p_value_apple_pre}")
print(f"Pre-COVID Samsung vs Tech Sector: T-statistic = {t_stat_samsung_pre}, P-value = {p_value_samsung_pre}")

```

```python
# Post-COVID t-tests

# Apple vs Tech Sector (QQQ)
t_stat_apple_post, p_value_apple_post = stats.ttest_ind(post_covid_returns['AAPL'], post_covid_returns['QQQ'])

# Samsung vs Tech Sector (QQQ)
t_stat_samsung_post, p_value_samsung_post = stats.ttest_ind(post_covid_returns['005930.KS'], post_covid_returns['QQQ'])

print(f"Post-COVID Apple vs Tech Sector: T-statistic = {t_stat_apple_post}, P-value = {p_value_apple_post}")
print(f"Post-COVID Samsung vs Tech Sector: T-statistic = {t_stat_samsung_post}, P-value = {p_value_samsung_post}")
```

```python
# Define pre- and post-COVID periods
pre_covid = returns.loc[:'2020-03-01']
post_covid = returns.loc['2020-03-01':]
```

```python
pre_covid_corr_apple = pre_covid['AAPL'].corr(pre_covid['QQQ'])
pre_covid_corr_samsung = pre_covid['005930.KS'].corr(pre_covid['QQQ'])

print(f"Pre-COVID Correlation (Apple vs Tech): {pre_covid_corr_apple}")
print(f"Pre-COVID Correlation (Samsung vs Tech): {pre_covid_corr_samsung}")
```

```python
post_covid_corr_apple = post_covid['AAPL'].corr(post_covid['QQQ'])
post_covid_corr_samsung = post_covid['005930.KS'].corr(post_covid['QQQ'])

print(f"Post-COVID Correlation (Apple vs Tech): {post_covid_corr_apple}")
print(f"Post-COVID Correlation (Samsung vs Tech): {post_covid_corr_samsung}")
```

```python
# Plot daily returns
plt.figure(figsize=(10,6))
plt.plot(returns['AAPL'], label='Apple')
plt.plot(returns['005930.KS'], label='Samsung')
plt.plot(returns['QQQ'], label='Tech Sector (QQQ)', linestyle='--')
plt.title('Daily Returns of Apple, Samsung, and Tech Sector')
plt.legend()
plt.show()
```
