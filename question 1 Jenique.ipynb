```python


import matplotlib.pyplot as plt
import pandas as pd

```

```python
#pull the data from apple and samsung 
ad =pd.read_csv('Apple data_cleaned.csv')
sd =pd.read_csv('Samsung data_cleaned.csv')

date_format = '%m/%d/%y'  

ad['Date'] = pd.to_datetime(ad['Date'],format=date_format)
sd['Date'] = pd.to_datetime(sd['Date'],format=date_format)

mdf = pd.merge(ad,sd, on='Date', how='inner') # x = apple,y = Samsung   
mdf.to_csv("merged_output.csv", index=False)

print(mdf)
```

```python
#Plot samsung close dates 
plt.figure(figsize=(20, 10))
plt.plot(mdf['Date'], mdf['Close_y'], marker='*', label='Samsung Closing Price', color='orange')
plt.title('Samsung Stock Prices Over Years 2017-2024')
plt.xlabel('Year')
plt.ylabel('Stock Price')
plt.axvline(pd.Timestamp('2020-01-01'), color='red', linestyle='solid', linewidth=3, label='Start of Covid')
plt.axvline(pd.Timestamp('2022-01-01'), color='blue', linestyle='solid', linewidth=3, label='After Covid')
plt.grid()
plt.legend()
plt.tight_layout()
plt.show()
```

```python
plt.savefig('Samsung Stock.png')
```

```python
#Plot Apple stock over years 2017-2024
plt.figure(figsize=(18, 10))
plt.plot(mdf['Date'], mdf['Close_x'], marker='.', label="Apple's Closing Price", color='orange')
plt.title('Apple Stock Prices Over Years 2017-2024')
plt.xlabel('Year')
plt.ylabel('Stock Price')
plt.axvline(pd.Timestamp('2020-01-01'), color='red', linestyle='solid', linewidth=3, label='Start of covid')
plt.axvline(pd.Timestamp('2022-01-01'), color='blue', linestyle='solid', linewidth=3, label='After Covid')
plt.grid()
plt.legend()
plt.tight_layout()
plt.show()
```
